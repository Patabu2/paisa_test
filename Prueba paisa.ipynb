```python
file = 'numbers.txt'

spanish_dict = {}

with open(file, 'r', encoding='utf-8') as numbers:
    for line in numbers:
        line_list = line.split(': ')
        
        integer = int(line_list[0])
        spanish_translation = line_list[1].strip('\n')
        
        spanish_dict[integer] = spanish_translation
```

```python
def int_to_text(integer):
    if integer <= 30:
        return spanish_dict[integer]
    
    
    int_str = str(integer)
    int_length = len(int_str)

    if int_length == 2:
        tenths = int_str[0]
        tenths += '0'
        
        if int_str[1] == '0':
            return spanish_dict[int(tenths)]

        return f'{spanish_dict[int(tenths)]} y {spanish_dict[int(int_str[1])]}'
    
    
    if int_length == 3:
        hundreds = int_str[0]
        hundreds += '00'
        
        tenths = int_str[1]
        tenths += '0'
        
        if int_str[-1] == '0':
            if int_str[1] == '0':
                return spanish_dict[int(hundreds)]
            
            
            if hundreds[0] == '1':
                return  f'{spanish_dict[int(hundreds)]}to {spanish_dict[int(tenths)]}'
            else:
                return  f'{spanish_dict[int(hundreds)]} {spanish_dict[int(tenths)]}'
        
        if hundreds[0] == '1':
            if int(int_str[1]) < 3:
                return f'{spanish_dict[int(hundreds)]}to {spanish_dict[int(int_str[-2:])]}'
            else:
                return f'{spanish_dict[int(hundreds)]}to {spanish_dict[int(tenths)]} y {spanish_dict[int(int_str[-1])]}'
        else:
            if int(int_str[1]) < 3:
                return f'{spanish_dict[int(hundreds)]} {spanish_dict[int(int_str[-2:])]}'
            else:
                return f'{spanish_dict[int(hundreds)]} {spanish_dict[int(tenths)]} y {spanish_dict[int(int_str[-1])]}'

        
    
int_to_text(100)
```

```python
for i in range(1000):
    print(int_to_text(i))
```
